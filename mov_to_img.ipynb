### iphoneの動画から画像へ変換

reverseをチェックすると180度画像を回転します。

```python
import cv2
import os

# @markdown ### Settings
root_path = 'drive/MyDrive'
project_folder = 'nvdiffrec_on_colab' #@param {type:"string"}
image_folder = 'hoge' #@param {type:"string"}
filename = 'hoge.mov' #@param {type:"string"}
reverse = True #@param {type:"boolean"}

def save_all_frames(video_path, dir_path, basename, ext='jpg'):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print('ファイルが開けません')
        return

    base_path = os.path.join(dir_path, basename)
    os.makedirs(base_path, exist_ok=True)

    digit = len(str(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))))
    n = 0
    while True:
        ret, frame = cap.read()
        if ret:
            if reverse:
                frame = cv2.rotate(frame, cv2.ROTATE_180)
            cv2.imwrite(f'{base_path}/image_{str(n).zfill(digit)}.{ext}', frame)
            n += 1
        else:
            return

save_all_frames(f'{root_path}/{project_folder}/{filename}', f'{root_path}/{project_folder}/{image_folder}', 'images')
```
